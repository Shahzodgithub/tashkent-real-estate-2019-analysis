```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

df = pd.read_excel('../input/tashkent-real-estate-2019/uybor.xlsx')
df.head(20)

# 1. **Qaysi hududda uylar narxi qimmat**

```python
plt.figure(figsize=(13,10))
sns.barplot(data=df, x='district', y='price')
plt.xticks(rotation=90)
plt.title("Uylar narxlari hududlar bo'yicha")
plt.show()
```

# 2. Xonalar soni va narxi o'rtasidagi korrelyatsiya

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x='rooms', y='price')
plt.title("Xonalar soni va narxi ")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='rooms', y='price', hue='level')
plt.show()
```

# 3. Xonalar o'lchami va narxi o'rtasidagi korrelyatsiya

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='size', y='price', hue='level')
sns.regplot(data=df, x='size', y='price')
plt.show()
```

# 4. Narxlar bo'yicha uylar soni

```python
plt.figure(figsize=(12,8))
sns.histplot(data=df, x='price')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(16,10))
sns.displot(data=df, x='price', height=5,aspect=2, hue='level', multiple="stack")
plt.xticks(rotation=90)
plt.show()
```

```python

sns.displot(data=df, x='lat',kind='kde', height=5)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(16,10))
sns.ecdfplot(data=df, x='size',hue='rooms')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,10))
sns.swarmplot(data=df, x='size', y='price')
plt.ylabel('Ulushi(%)')
plt.xticks(rotation=90)
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(15,10))
fig.suptitle("Grafik 1")
sns.histplot(ax=ax[0, 0], data=df, x='lat',y='price')
sns.histplot(ax=ax[0, 1], data=df, x='size')
sns.histplot(ax=ax[1, 0], data=df, x='level')
sns.histplot(ax=ax[1, 1], data=df, x='lat')
plt.show()
```
